```python
import pandas as pd  ## Libreria pandas
import numpy as np   ## Libreria numpy para estadistica
from pathlib import Path ## Libreria pathlib para trabajar en rutas

pd.set_option("display.max_columns", None) ## Mostrar todas las columnas
```

```python
## Variables para guardar la ubicación de los archivos

comtrade_path = Path("../data/raw/un_comtrade_coffee_cocoa_2020_2025.csv")
population_path = Path("../data/raw/worldbank_population/wb_population_data.csv")
gdp_path = Path("../data/raw/worldbank_gdp/wb_gdp_per_capita_data.csv")
```

```python
## Comprobación que la ruta de los archivos es correcta y que existen los archivos

print("Comtrade exists:", comtrade_path.exists())
print("Population exists:", population_path.exists())
print("GDP exists:", gdp_path.exists())
```

```python
# Carga de los datasets procedentes de las fuentes oficiales.
# UN Comtrade proporciona los datos de comercio internacional.
# World Bank proporciona los indicadores demográficos y económicos.
#
# Los archivos de World Bank contienen varias líneas iniciales de metadatos
# que no forman parte de la tabla de datos, por lo que se omiten mediante
# el parámetro skiprows=4.
#
# El archivo de UN Comtrade utiliza una codificación distinta de UTF-8,
# por lo que se especifica explícitamente la codificación latin1 para evitar
# errores de lectura.
#
# Además, el CSV de UN Comtrade contiene una coma adicional al final de cada fila.
# Para evitar que Pandas interprete incorrectamente la primera columna como índice
# y desplace el resto de columnas, se utiliza index_col=False.

comtrade = pd.read_csv(
    comtrade_path,
    encoding="latin1",
    index_col=False
)
population = pd.read_csv(population_path, skiprows=4)
gdp = pd.read_csv(gdp_path, skiprows=4)
```

```python
# Inspección inicial de los datasets.
#
# Este paso forma parte de la validación inicial de los datos
# antes de realizar cualquier limpieza o transformación.
#
# shape devuelve el número de filas y columnas de cada tabla,
# permitiendo comprobar que los archivos se han cargado correctamente
# y conocer su tamaño antes de comenzar el análisis.
#
# head() muestra las primeras filas de cada dataset para revisar:
# - la estructura de las columnas
# - el tipo de información disponible
# - posibles problemas de carga o formato
# - la presencia de valores nulos o columnas innecesarias


print("Comtrade:", comtrade.shape)
print("Population:", population.shape)
print("GDP:", gdp.shape)

display(comtrade.head())
display(population.head())
display(gdp.head())
```

```python
# Revisión técnica de la estructura del dataset de comercio internacional.
# info() muestra el número de filas, columnas, tipos de datos y valores no nulos.
# Este paso permite identificar qué variables son numéricas, cuáles son categóricas 
# y qué columnas pueden contener valores nulos o no aportar información útil.

comtrade.info()
```

```python
# Revisión técnica del dataset de población.
# El objetivo es comprobar su estructura, el número de registros disponibles y la presencia de valores nulos por año.
# Esta revisión confirma que 2025 no tiene datos disponibles y que existe
# una columna vacía generada por la descarga: Unnamed: 70.

population.info()
```

```python
# Revisión técnica del dataset de PIB per cápita.
#
# info() permite comprobar la estructura del dataset, los tipos de datos
# y la disponibilidad de información por año.
#
# Esta revisión servirá para identificar posibles valores nulos y determinar
# qué años y países pueden utilizarse posteriormente en el análisis.

gdp.info()
```

```python
# Validación de las variables principales del dataset de comercio.
#
# Se revisan las columnas que serán utilizadas posteriormente en el análisis:
# - País importador
# - Flujo comercial
# - Producto comercializado
# - Valor de las importaciones
#
# Esta comprobación confirma que el archivo se ha cargado correctamente
# y que las variables clave contienen información coherente.

comtrade[
    [
        "reporterDesc",
        "flowDesc",
        "partnerDesc",
        "cmdDesc",
        "primaryValue"
    ]
].head()
```

```python
# Revisión de los productos incluidos en el dataset de comercio.
#
# El objetivo es comprobar qué códigos HS se han descargado
# y cuántos registros existen para cada producto.
#
# Esta validación permite confirmar que el dataset contiene café (HS 0901)
# cacao en grano (HS 1801) y preparados de cacao (HS 1806) antes de iniciar la limpieza.

comtrade["cmdCode"].value_counts()
```

```python
# Revisión de las descripciones de producto asociadas a los códigos HS.
#
# Esta comprobación ayuda a interpretar correctamente cada código
# y a construir posteriormente una variable de negocio más legible,
# como product_group o sector.

comtrade["cmdDesc"].value_counts()
```

```python
# Revisión de los años disponibles en el dataset de comercio.
#
# Esta comprobación permite validar si el periodo descargado contiene
# datos entre 2020 y 2025 y si la cobertura por año es equilibrada.
#
# Es importante porque los datos de importaciones podrán analizarse hasta 2025,
# mientras que los indicadores de World Bank solo estarán disponibles hasta 2024.

comtrade["refYear"].value_counts().sort_index()

# 2025 tiene 205 registros frente a los +/- 400 de los años anteriores, lo que indica que la cobertura de datos para ese año es limitada.
```

```python
# Revisión del número de países importadores disponibles en el dataset.
#
# Esta comprobación permite conocer la cobertura geográfica inicial
# antes de filtrar el análisis a países de la Unión Europea.

print(comtrade["reporterDesc"].nunique())
```

```python
# Listado de países importadores incluidos en el dataset.
#
# El dataset descargado contiene más países que los de la Unión Europea,
# por lo que posteriormente será necesario filtrar únicamente los países
# relevantes para el análisis.

sorted(comtrade["reporterDesc"].unique())
```

```python
# Definición de países de la Unión Europea.
#
# El dataset original contiene países de todo el mundo.
# Como el objetivo del proyecto es analizar oportunidades en mercados europeos
# bajo un marco regulatorio comparable, se filtra el análisis a países miembros
# de la Unión Europea.

eu_countries = [
    "Austria",
    "Belgium",
    "Bulgaria",
    "Croatia",
    "Cyprus",
    "Czechia",
    "Denmark",
    "Estonia",
    "Finland",
    "France",
    "Germany",
    "Greece",
    "Hungary",
    "Ireland",
    "Italy",
    "Latvia",
    "Lithuania",
    "Luxembourg",
    "Malta",
    "Netherlands",
    "Poland",
    "Portugal",
    "Romania",
    "Slovakia",
    "Slovenia",
    "Spain",
    "Sweden"
]
```

```python
# Filtrado del dataset de comercio para conservar únicamente países de la Unión Europea.
#
# Se crea una copia del dataframe filtrado para evitar modificar el dataset original
# descargado desde UN Comtrade.

comtrade_eu = comtrade[
    comtrade["reporterDesc"].isin(eu_countries)
].copy()

print("Dataset original:", comtrade.shape)
print("Dataset UE:", comtrade_eu.shape)
print("Países UE incluidos:", comtrade_eu["reporterDesc"].nunique())
```

```python
# Revisión de la cobertura temporal después de filtrar países de la Unión Europea.
#
# Esta comprobación permite ver cuántos registros existen por año
# dentro del mercado objetivo del análisis.

comtrade_eu["refYear"].value_counts().sort_index()
```

# El filtrado a países de la Unión Europea deja 471 registros.
# Para el periodo 2020-2024 la cobertura es completa:
# 27 países x 3 productos = 81 registros por año.
#
# En 2025 solo existen 66 registros, por lo que ese año parece incompleto
# en el momento de la descarga. Por este motivo, las métricas comparativas
# principales se calcularán sobre 2020-2024, mientras que 2025 podrá utilizarse
# únicamente como dato reciente con advertencia metodológica.

```python
# Selección de columnas útiles para el análisis.
#
# El dataset original de UN Comtrade contiene muchas columnas técnicas
# que no son necesarias para responder a la pregunta de negocio.
# En esta fase se conservan únicamente las variables necesarias para:
# - identificar país y año
# - identificar producto
# - analizar valor importado
# - analizar volumen importado

trade_clean = comtrade_eu[
    [
        "refYear",
        "reporterISO",
        "reporterDesc",
        "cmdCode",
        "cmdDesc",
        "netWgt",
        "primaryValue"
    ]
].copy()

trade_clean.head()
```

```python
# Renombrado de columnas.
#
# Se sustituyen los nombres técnicos originales por nombres más claros
# y orientados al análisis de negocio.

trade_clean = trade_clean.rename(
    columns={
        "refYear": "year",
        "reporterISO": "country_code",
        "reporterDesc": "country",
        "cmdCode": "commodity_code",
        "cmdDesc": "commodity_description",
        "netWgt": "net_weight_kg",
        "primaryValue": "import_value_usd"
    }
)

trade_clean.head()
```

```python
# Creación de variables de negocio.
#
# Los códigos HS son útiles técnicamente, pero no son intuitivos para una audiencia no técnica.
# Por eso se crean columnas más legibles:
# - sector: diferencia entre café y cacao
# - product_group: clasifica cada producto en una categoría de negocio
#
# Estas variables facilitarán la construcción de filtros y gráficos en el dashboard.

product_mapping = {
    901: {
        "sector": "Coffee",
        "product_group": "Coffee total"
    },
    1801: {
        "sector": "Cocoa",
        "product_group": "Cocoa beans"
    },
    1806: {
        "sector": "Cocoa",
        "product_group": "Chocolate and cocoa preparations"
    }
}

trade_clean["sector"] = trade_clean["commodity_code"].map(
    lambda x: product_mapping[x]["sector"]
)

trade_clean["product_group"] = trade_clean["commodity_code"].map(
    lambda x: product_mapping[x]["product_group"]
)

trade_clean.head()
```

```python
# Validación de las nuevas variables de negocio.
#
# Se comprueba que cada código de producto se haya clasificado correctamente
# en su sector y grupo de producto correspondiente.

trade_clean[
    ["commodity_code", "sector", "product_group"]
].drop_duplicates().sort_values("commodity_code")
```

```python
# Cálculo del precio medio por kilogramo.
#
# Esta métrica permite comparar el valor económico relativo de cada producto
# más allá del volumen importado.
#
# Se calcula dividiendo el valor total importado entre el peso neto importado.
# Si el peso neto es cero, el resultado se deja como valor nulo para evitar
# divisiones incorrectas.

trade_clean["avg_price_usd_kg"] = np.where(
    trade_clean["net_weight_kg"] > 0,
    trade_clean["import_value_usd"] / trade_clean["net_weight_kg"],
    np.nan
)

trade_clean.head()
```

```python
# Validación estadística inicial de las principales métricas numéricas.
#
# describe() permite revisar mínimos, máximos, medias y dispersión.
# Este paso ayuda a detectar valores extremos, nulos o precios medios poco realistas que puedan requerir limpieza.

trade_clean[
    ["net_weight_kg", "import_value_usd", "avg_price_usd_kg"]
].describe()
```

```python
# Identificación de registros nulos o igual a cero.
#
# Estos registros impiden calcular el precio medio por kilogramo y pueden
# afectar a determinados análisis comparativos.
#
# La revisión permitirá decidir posteriormente si deben eliminarse,
# imputarse o mantenerse únicamente para análisis de valor comercial.

trade_clean[
    trade_clean["net_weight_kg"].isna() |
    (trade_clean["net_weight_kg"] == 0)
][
    [
        "year",
        "country",
        "commodity_code",
        "product_group",
        "net_weight_kg",
        "import_value_usd"
    ]
]
```

# Los registros sin peso neto o con peso igual a cero no permiten calcular
# correctamente el precio medio por kilogramo.
#
# Sin embargo, estos registros sí contienen valor de importación, por lo que
# se mantienen en el dataset para análisis de valor comercial.
#
# La variable avg_price_usd_kg permanecerá como valor nulo en estos casos
# y estos registros se excluirán únicamente de análisis basados en precio/kg.

```python
# Revisión de posibles duplicados.
#
# La unidad de análisis esperada es país, año y producto.
# Por tanto, no debería haber más de un registro para la misma combinación
# de country_code, year y commodity_code.

duplicates = trade_clean.duplicated(
    subset=["country_code", "year", "commodity_code"],
    keep=False
)

print("Número de registros duplicados:", duplicates.sum())

trade_clean[duplicates].sort_values(
    ["country", "year", "commodity_code"]
)
```

```python
# Limpieza inicial del dataset de población.
#
# Se elimina la columna vacía generada durante la descarga
# y el año 2025, ya que no contiene datos.
#
# Esta limpieza simplifica las transformaciones posteriores.

population = population.drop(
    columns=["Unnamed: 70", "2025"]
)
```

```python
# Limpieza inicial del dataset de PIB per cápita.
#
# Se elimina la columna vacía generada durante la descarga
# y el año 2025, ya que no contiene datos disponibles.

gdp = gdp.drop(
    columns=["Unnamed: 70", "2025"]
)
```

```python
# Transformación del dataset de población trasponiendo las columnas de año.
#
# El dataset original está en formato ancho (un año por columna).
# Para poder relacionarlo con el dataset de comercio,
# se transforma a formato largo (el año es una columna con valores) utilizando melt().
#
# El resultado tendrá una fila por país y año.

population_long = population.melt(
    id_vars=[
        "Country Name",
        "Country Code",
        "Indicator Name",
        "Indicator Code"
    ],
    var_name="year",
    value_name="population"
)

population_long.head()
```

```python
# Transformación del dataset de PIB per cápita a formato largo.
#
# Igual que en población, cada año pasa de ser una columna
# a convertirse en un registro independiente.

gdp_long = gdp.melt(
    id_vars=[
        "Country Name",
        "Country Code",
        "Indicator Name",
        "Indicator Code"
    ],
    var_name="year",
    value_name="gdp_per_capita"
)

gdp_long.head()
```

```python
# Validación de la transformación a formato largo.
print("Population long:", population_long.shape)
print("GDP long:", gdp_long.shape)

display(population_long.head())
display(gdp_long.head())
```

```python
# Filtrado de los datasets auxiliares.
#
# Se conservan únicamente los países de la Unión Europea y los años necesarios para cruzar con el dataset de comercio.
# En World Bank, el año queda como texto tras melt(), por lo que se convierte a entero.

population_long["year"] = population_long["year"].astype(int)
gdp_long["year"] = gdp_long["year"].astype(int)

population_eu = population_long[
    (population_long["Country Name"].isin(eu_countries)) &
    (population_long["year"].between(2020, 2024))
].copy()

gdp_eu = gdp_long[
    (gdp_long["Country Name"].isin(eu_countries)) &
    (gdp_long["year"].between(2020, 2024))
].copy()

print("Population EU:", population_eu.shape)
print("GDP EU:", gdp_eu.shape)
```

```python
# Comprobación de países de la Unión Europea no encontrados en World Bank.
#
# El objetivo es detectar diferencias de nomenclatura entre fuentes,
# por ejemplo "Czechia" frente a otro nombre equivalente.

wb_countries = set(population_long["Country Name"].unique())
missing_countries = sorted(set(eu_countries) - wb_countries)

missing_countries
```

```python
# Normalización de nombres de países entre fuentes.
#
# UN Comtrade y World Bank no siempre utilizan exactamente la misma
# nomenclatura para los países. En este caso, World Bank usa
# "Slovak Republic" mientras que UN Comtrade usa "Slovakia".
#
# Se corrige el nombre para poder cruzar correctamente ambos datasets.

country_name_mapping = {
    "Slovak Republic": "Slovakia"
}

population_long["Country Name"] = population_long["Country Name"].replace(country_name_mapping)
gdp_long["Country Name"] = gdp_long["Country Name"].replace(country_name_mapping)
```

```python
# Selección y renombrado de columnas de población.
#
# Se conservan únicamente las variables necesarias para el modelo analítico.
# El resultado será una tabla dimensional relacionada por país y año.

population_final = population_eu[
    [
        "Country Code",
        "Country Name",
        "year",
        "population"
    ]
].copy()

population_final = population_final.rename(
    columns={
        "Country Code": "country_code",
        "Country Name": "country"
    }
)

population_final.head()
```

```python
# Selección y renombrado de columnas de PIB per cápita.
#
# Se eliminan las columnas descriptivas que ya no son necesarias
# para el análisis y se adopta la misma nomenclatura utilizada
# en el resto del proyecto.

gdp_final = gdp_eu[
    [
        "Country Code",
        "Country Name",
        "year",
        "gdp_per_capita"
    ]
].copy()

gdp_final = gdp_final.rename(
    columns={
        "Country Code": "country_code",
        "Country Name": "country"
    }
)

gdp_final.head()
```

```python
# Validación final de los datasets limpios y transformados.
#
# Se comprueba que los datasets de comercio, población y PIB per cápita
# tienen el número esperado de registros y columnas antes de iniciar
# el análisis exploratorio y la construcción del modelo analítico.

print(trade_clean.shape)
print(population_final.shape)
print(gdp_final.shape)
```

```python
# Integración de comercio con población.
#
# Se cruza el dataset de importaciones con la población utilizando
# país y año como claves comunes.
#
# Se usa left join para conservar todos los registros de comercio
# y añadir población cuando esté disponible.

trade_enriched = trade_clean.merge(
    population_final,
    on=["country_code", "country", "year"],
    how="left"
)

trade_enriched.head()
```

```python
# Integración del PIB per cápita.
#
# Se añade el indicador económico al dataset enriquecido.
# Al igual que en el cruce anterior, se conserva la tabla de comercio
# como base principal del análisis.

trade_enriched = trade_enriched.merge(
    gdp_final,
    on=["country_code", "country", "year"],
    how="left"
)

trade_enriched.head()
```

```python
# Validación del dataset integrado.
#
# Se revisa el tamaño final y la presencia de valores nulos
# en las variables añadidas desde World Bank.

print(trade_enriched.shape)

trade_enriched[
    ["year", "country", "product_group", "population", "gdp_per_capita"]
].isna().sum()
```

```python
# Comprobación de los valores nulos tras integrar World Bank.
#
# La ausencia de población y PIB per cápita debería corresponder al año 2025,
# ya que los datasets de World Bank descargados no contienen valores para ese año.

trade_enriched[
    trade_enriched["population"].isna()
]["year"].value_counts()
```

```python
# Creación del dataset enriquecido para métricas per cápita.
#
# Se excluye 2025 porque no dispone de población ni PIB per cápita.
# El año 2025 se mantendrá disponible para análisis descriptivo de importaciones,
# pero no para métricas cruzadas con indicadores socioeconómicos.

trade_enriched_complete = trade_enriched[
    trade_enriched["year"].between(2020, 2024)
].copy()

trade_enriched_complete[
    ["population", "gdp_per_capita"]
].isna().sum()
```

```python
# Cálculo de métricas per cápita.
#
# Estas variables permiten comparar países de distinto tamaño poblacional.
# Un país grande puede importar más en valor absoluto, pero no necesariamente
# tener mayor intensidad de mercado por habitante.
#
# Se calculan únicamente sobre el dataset 2020-2024, ya que 2025 no dispone
# de población ni PIB per cápita en World Bank.

trade_enriched_complete["import_value_usd_per_capita"] = (
    trade_enriched_complete["import_value_usd"] / trade_enriched_complete["population"]
)

trade_enriched_complete["net_weight_kg_per_capita"] = (
    trade_enriched_complete["net_weight_kg"] / trade_enriched_complete["population"]
)

trade_enriched_complete.head()
```

```python
# Validación estadística de las métricas per cápita.
#
# Se revisan las principales métricas calculadas para detectar valores extremos,
# nulos o resultados poco coherentes antes de utilizarlas en el dashboard.

trade_enriched_complete[
    [
        "import_value_usd_per_capita",
        "net_weight_kg_per_capita",
        "gdp_per_capita"
    ]
].describe()
```

```python
trade_enriched_complete.columns.tolist()
```

```python
# Limpieza de columnas duplicadas tras los cruces.
#
# Durante el proceso de integración se generaron columnas duplicadas
# de PIB per cápita con sufijos _x e _y.
# Se conserva una única variable gdp_per_capita para evitar ambigüedades
# en el dataset final.

trade_enriched_complete = trade_enriched_complete.drop(
    columns=["gdp_per_capita_x", "gdp_per_capita_y"]
)

trade_enriched_complete.columns.tolist()
```

```python
# Validación final de nulos en el dataset enriquecido.
#
# Esta revisión permite comprobar qué variables conservan valores nulos
# antes de exportar el dataset procesado.

trade_enriched_complete.isna().sum()
```

```python
# Exportación del dataset procesado.
#
# Se guarda el dataset final limpio en la carpeta data/processed.
# Este archivo será la base para el dashboard de Streamlit.

processed_path = Path("../data/processed/trade_enriched_complete_2020_2024.csv")

trade_enriched_complete.to_csv(
    processed_path,
    index=False
)

print("Archivo exportado en:", processed_path)
```
